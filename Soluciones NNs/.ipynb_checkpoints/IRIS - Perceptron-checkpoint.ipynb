```python
# Import the relevant packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.colors import ListedColormap # For the colored decision regions
```

### Función para dibujar las regiones de los datos 

```python
def plot_decision_regions(X, y, classifier, resolution=0.02):
    # setup marker generator and color map
    markers = ('o', 'x', 's', '^', 'v')
    colors = ('r', 'b', 'g', 'k', 'grey')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision regions by creating a pair of grid arrays xx1 and xx2 via meshgrid function in Numpy
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    
    # use predict method to predict the class labels z of the grid points
    Z = classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    # draw the contour using matplotlib
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # plot class samples
    for i, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl, 0], y=X[y==cl, 1], alpha=0.8, c=cmap(i), marker=markers[i], label=cl)
```

# Código para el perceptrón

El perceptrón presente aquí es una versión simple para clasificación, hecha a mano sin emplear funciones predefinidas. Puede hacerse más facilmente empleando __scipy__ 

```python
# Build the perceptron as an object

class Perceptron(object):
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
        Learning rate (between 0.0 and 1.0)
    n_iter : int
        Passes over the training dataset.
    mean_bias_init: boolean
        Checks whether to initialize the bias at the mean of the second variable

    Attributes
    -----------
    w_ : 1d-array
        Weights after fitting.
    errors_ : list
        Number of misclassifications in every epoch.

    """
    def __init__(self, eta = 1.0, n_iter = 100, mean_bias_init = False):
        self.eta = eta
        self.n_iter = n_iter
        self.mean_bias_init = mean_bias_init
    
    
    def fit(self, X, y, weight_init = []):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
            Training vectors, where n_samples is the number of samples and
            n_features is the number of features.
        y : array-like, shape = [n_samples]
            Target values.
        weight_init : list
            Passes an initialization for the weights
            
        Returns
        -------
        self : object

        """
        # Initialize weights as zeros of size 1 + number of features, errors as empty list
        if len( weight_init ) == 0:
            self.w_ = np.zeros(1 + X.shape[1])  # Add the extra dimension to make room for the bias component
            if (self.mean_bias_init):
                self.w_[0] = np.mean(X[:,1])
        else:
             self.w_ = np.array(weight_init)   
        
        self.errors_ = []                   # Empty list to store the errors during training 
        
        for _ in range(self.n_iter):
            
            # Initialize the 0-1 loss counter
            errors = 0
            
            # Iterate samples one by one and update the weights
            for xi, target in zip(X, y):

                yhat = self.predict(xi)
                error_condition = target * yhat
                
                if (error_condition <= 0):
                    self.w_[0] += self.eta * target
                    self.w_[1:] += self.eta * target * xi
                    errors += 1
            
            self.errors_.append(errors)
            
            if (errors == 0.0):
                break
        
        return self
    
    def net_input(self, X):
        """Calculate net input before activation"""
        return np.dot(X, self.w_[1:]) + self.w_[0]        # Linear transformation of the input
    
    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1)  # Sign function in compact form
```

## Datos de IRIS 

Más info: https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html

```python
# Load and read the "IRIS" dataset (default from python)
from sklearn.datasets import load_iris

data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df["Species"] = data.target

df.head()
```

Uasmos la información presente online sobre el dataset

```python
# Select the two classes
y = df.iloc[0:100, 4].values  # values method of a pandas dataframe yields Numpy array
y = np.where(y == 0, -1, 1) # Setosa as (-1), versicolor as (+1)

# select two features: sepal length and petal length for visualization
X = df.iloc[0:100, [0,2]].values
```

Dibujamos los puntos que nos quedan

```python
plt.scatter(X[:50, 0], X[:50, 1], color='darkred', marker='o', label='sentosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='darkblue', marker='x', label='versicolor')
plt.xlabel('Sepal length [cm]', fontsize = 20)
plt.ylabel('Petal length [cm]', fontsize = 20)
plt.legend(loc='upper left', fontsize = 20)
plt.show()
```

# Entrenamiento y plots de resultados

```python
# First, create the perceptron object
ppn = Perceptron(eta = 1.0, n_iter = 1000, mean_bias_init = True)

ppn.fit(X, y)

# plot the error for each epoch to check for convergence
plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_)
plt.xlabel('Epochs', fontsize = 20)
plt.ylabel('Number of errors',  fontsize = 20)
plt.show()
```

```python
plot_decision_regions(X, y, ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```

## Sin inicialización 

```python
ppn_nobias = Perceptron(eta = 1.0, n_iter = 1000, mean_bias_init = False)
ppn_nobias.fit(X, y)

f = plt.figure(figsize=(20, 4))
f.add_subplot(1,2, 1)
plot_decision_regions(X, y, ppn_nobias)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')

f.add_subplot(1,2,2)
plt.plot(range(1, len(ppn_nobias.errors_)+1), ppn_nobias.errors_)
plt.xlabel('Epochs', fontsize = 20)
plt.ylabel('Number of errors',  fontsize = 20)

plt.show()
```

## Con inicialización dada 

Damos los parámetros en _weight_init_

```python
# Initialization weight vector 
weight_init = [1.0, 1000, 0]    # Start with an almost vertical line

ppn_weights = Perceptron(eta = 1.0, n_iter = 1000)
ppn_weights.fit(X, y, weight_init = weight_init)

f = plt.figure(figsize=(20, 4))
f.add_subplot(1,2, 1)
plot_decision_regions(X, y, ppn_weights)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')

f.add_subplot(1,2,2)
plt.plot(range(1, len(ppn_weights.errors_)+1), ppn_weights.errors_)
plt.xlabel('Epochs', fontsize = 20)
plt.ylabel('Number of errors',  fontsize = 20)

plt.show()
```

## *Extra*: Añadimos ruido a los datos

```python
# Add gaussian noise to the X positions
np.random.seed(123)
X_noisy = X + np.random.normal(0, 0.2, X.shape)

# Plot the data to check whether it is still linearly separable
plt.scatter(X_noisy[:50, 0], X_noisy[:50, 1], color='darkred', marker='o', label='sentosa')
plt.scatter(X_noisy[50:100, 0], X_noisy[50:100, 1], color='darkblue', marker='x', label='versicolor')
plt.xlabel('Sepal length [cm]', fontsize = 20)
plt.ylabel('Petal length [cm]', fontsize = 20)
plt.legend(loc='upper left', fontsize = 20)
plt.show()
```

```python
# Relaunch the perceptron in this case

ppn = Perceptron(eta = 1.0, n_iter = 1000) 
ppn.fit(X_noisy, y)

f = plt.figure(figsize=(20, 4))
f.add_subplot(1,2, 1)
plot_decision_regions(X_noisy, y, ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')

f.add_subplot(1,2,2)
plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_)
plt.xlabel('Epochs', fontsize = 20)
plt.ylabel('Number of errors',  fontsize = 20)

plt.show()
```

### ¿Qué pasa con aún más ruido?

```python
# Add gaussian noise to the X positions
np.random.seed(123)
X_noisy = X + np.random.normal(0, 0.5, X.shape)

# Plot the data to check whether it is still linearly separable
plt.scatter(X_noisy[:50, 0], X_noisy[:50, 1], color='darkred', marker='o', label='sentosa')
plt.scatter(X_noisy[50:100, 0], X_noisy[50:100, 1], color='darkblue', marker='x', label='versicolor')
plt.xlabel('Sepal length [cm]', fontsize = 20)
plt.ylabel('Petal length [cm]', fontsize = 20)
plt.legend(loc='upper left', fontsize = 20)
plt.show()
```

**Los datos ahora no son linealmente separables, así que el perceptrón no debería terminar su entrenamiento nunca** (especificamos un número máximo de iteraciones para cortarlo)

```python
# Relaunch the perceptron in this case

ppn = Perceptron(eta = 1.0, n_iter = 1000) 
ppn.fit(X_noisy, y)

f = plt.figure(figsize=(20, 4))
f.add_subplot(1,2, 1)
plot_decision_regions(X_noisy, y, ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')

f.add_subplot(1,2,2)
plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_)
plt.xlabel('Epochs', fontsize = 20)
plt.ylabel('Number of errors',  fontsize = 20)

plt.show()
```
